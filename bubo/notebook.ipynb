```python
# from transition import *
from easing import Linear, Ease_in_quad, Ease_out_quad, Ease_in_out_quad, Ease_in_cubic, Ease_out_cubic, Ease_in_out_cubic, Ease_in_quart, Ease_out_quart, Ease_in_out_quart, Ease_in_quint, Ease_out_quint, Ease_in_out_quint, Ease_in_sine, Ease_out_sine, Ease_in_out_sine, Ease_in_expo, Ease_out_expo, Ease_in_out_expo, Ease_in_circ, Ease_out_circ, Ease_in_out_circ, Ease_in_elastic, Ease_out_elastic, Ease_in_out_elastic, Ease_in_back, Ease_out_back, Ease_in_out_back

import numpy as np
from matplotlib import pyplot as plt
```

```python
linear_tween = Linear(start=3, end=1, duration=1)

ease_in_quad = Ease_in_quad(start=3, end=1, duration=1)
c = Ease_out_quad(start=3, end=1, duration=1)
d = Ease_in_out_quad(start=3, end=1, duration=1)

ease_in_cubic = Ease_in_cubic(start=3, end=1, duration=1)
f = Ease_out_cubic(start=3, end=1, duration=1)
g = Ease_in_out_cubic(start=3, end=1, duration=1)

ease_in_quart = Ease_in_quart(start=3, end=1, duration=1)
i = Ease_out_quart(start=3, end=1, duration=1)
j = Ease_in_out_quart(start=3, end=1, duration=1)

ease_in_quint = Ease_in_quint(start=3, end=1, duration=1)
l = Ease_out_quint(start=3, end=1, duration=1)
m = Ease_in_out_quint(start=3, end=1, duration=1)

ease_in_sine = Ease_in_sine(start=3, end=1, duration=1)
o = Ease_out_sine(start=3, end=1, duration=1)
p = Ease_in_out_sine(start=3, end=1, duration=1)

ease_in_expo = Ease_in_expo(start=3, end=1, duration=1)
r = Ease_out_expo(start=3, end=1, duration=1)
s = Ease_in_out_expo(start=3, end=1, duration=1)

ease_in_circ = Ease_in_circ(start=3, end=1, duration=1)
u = Ease_out_circ(start=3, end=1, duration=1)
v = Ease_in_out_circ(start=3, end=1, duration=1)

ease_in_elastic = Ease_in_elastic(start=3, end=1, duration=1)
xx = Ease_out_elastic(start=3, end=1, duration=1)
y = Ease_in_out_elastic(start=3, end=1, duration=1)

ease_in_back = Ease_in_back(start=3, end=1, duration=1)
aa = Ease_out_back(start=3, end=1, duration=1)
bb = Ease_in_out_back(start=3, end=1, duration=1)

x = np.arange(0,1,0.001)
```

```python
# Generate data for our charts

x1 = list(map(linear_tween,x))
ease_in_quad_x = list(map(ease_in_quad,x))
x3 = list(map(c,x))
x4 = list(map(d,x))

ease_in_cubic_x = list(map(ease_in_cubic,x))
x6 = list(map(f,x))
x7 = list(map(g,x))

ease_in_quart_x = list(map(ease_in_quart,x))
x9 = list(map(i,x))
x10 = list(map(j,x))

ease_in_quint_x = list(map(ease_in_quint,x))
x12 = list(map(l,x))
x13 = list(map(m,x))

ease_in_sine_x = list(map(ease_in_sine,x))
x15 = list(map(o,x))
x16 = list(map(p,x))

ease_in_expo_x = list(map(ease_in_expo,x))
x18 = list(map(r,x))
x19 = list(map(s,x))

ease_in_circ_x = list(map(ease_in_circ,x))
x21 = list(map(u,x))
x22 = list(map(v,x))

ease_in_elastic_x = list(map(ease_in_elastic,x))
x24 = list(map(xx,x))
x25 = list(map(y,x))

ease_in_back_x = list(map(ease_in_back,x))
x27 = list(map(aa,x))
x28 = list(map(bb,x))
```

```python
plt.plot(x,x1, label='Linear')
plt.legend()
plt.suptitle('Linear Easing')
```

```python


plt.plot(x,ease_in_quad_x, label='Ease_in_quad')
plt.plot(x,x3, label='Ease_out_quad')
plt.plot(x,x4, label='Ease_in_out_quad')
plt.legend()
plt.suptitle('Quad Easing')
```

```python
plt.plot(x,ease_in_cubic_x, label='Ease_in_cubic')
plt.plot(x,x6, label='Ease_out_cubic')
plt.plot(x,x7, label='Ease_in_out_cubic')
plt.legend()
plt.suptitle('Cubic Easing')
```

```python
plt.plot(x,ease_in_quart_x, label='Ease_in_quart')
plt.plot(x,x9, label='Ease_out_quart')
plt.plot(x,x10, label='Ease_in_out_quart')
plt.legend()
plt.suptitle('Quart Easing')
```

```python
plt.plot(x,ease_in_quint_x, label='Ease_in_quint')
plt.plot(x,x12, label='Ease_out_quint')
plt.plot(x,x13, label='Ease_in_out_quint')
plt.legend()
plt.suptitle('Quint Easing')
```

```python
plt.plot(x,ease_in_sine_x, label='Ease_in_sine')
plt.plot(x,x15, label='Ease_out_sine')
plt.plot(x,x16, label='Ease_in_out_sine')
plt.legend()
plt.suptitle('Sine Easing')
```

```python
plt.plot(x,ease_in_expo_x, label='Ease_in_expo')
plt.plot(x,x18, label='Ease_out_expo')
plt.plot(x,x19, label='Ease_in_out_expo')
plt.legend()
plt.suptitle('Expo Easing')
```

```python
plt.plot(x,ease_in_circ_x, label='Ease_in_circ')
plt.plot(x,x21, label='Ease_out_circ')
plt.plot(x,x22, label='Ease_in_out_circ')
plt.legend()
plt.suptitle('Circ Easing')
```

```python
plt.plot(x,ease_in_elastic_x, label='Ease_in_elastic')
plt.plot(x,x24, label='Ease_out_elastic')
plt.plot(x,x25, label='Ease_in_out_elastic')
plt.legend()
plt.suptitle('Elastic Easing')
```
